```python
from math import sqrt, pi, exp
import numpy as np
import matplotlib.pyplot as plt
```

```python
domaine = range(-100, 100)
avg = 0
sigma = 20

# Normal distribution or Gaussian distribution
f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-avg),2)/(2*pow(sigma,2)))

y = [f(x) for x in domaine]
plot = plt.plot(domaine, y)
```

```python
randon_matrix_sample = np.random.rand(100, 200)
#sum_matrix = np.sum(randon_matrix_sample, 1) # rows sum
sum_matrix = np.sum(randon_matrix_sample, 0) # columns sum
print("The variable size is {}.".format(sum_matrix.shape))
```

```python
plot = plt.scatter(range(200), sum_matrix)
```

```python
plot = plt.hist(sum_matrix)
```

```python
random_matrix = np.random.rand(10000,10000)
sum_matrix = np.sum(random_matrix, 0)
plot = plt.hist(sum_matrix, bins=100)
```

```python
print("The empirical mean of our distribution is {}" 
        .format(np.mean(sum_matrix)))
print("The empirical mean of the variable generated by the function rand is {}" 
        .format(np.mean(np.random.rand(10000))))
print("The empirical variance of of our distribution is {}" 
        .format(np.var(sum_matrix)))
print("The empirical variance of the variable generated by the function rand is {}" 
        .format(np.var(np.random.rand(10000))))
```
